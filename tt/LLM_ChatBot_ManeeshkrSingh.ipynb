```python
# for language model
import transformers
import datetime
import numpy as np
import os
```

```python
from textblob import TextBlob

def get_feedback_score(input_text):
    blob = TextBlob(input_text)
    score = blob.sentiment.polarity
    #print(f"Feedback: {input_text}")
    #print(f"Score: {score}")
    return score

def get_feedback_level(score):
    if score >= 0.7:
        return 'Very Satisfied'
    elif score >= 0.2:
        return 'Satisfied'
    elif score >= 0.0:
        return 'Neutral'
    elif score >= -0.555:
        return 'Dissatisfied '
    else:
        return 'Very Dissatisfied'
```

```python
from transformers import pipeline, Conversation
# Create a sentiment analysis pipeline
nlp = pipeline("sentiment-analysis")
list_of_questions = [
    "How was the food and your overall experience?",
    "Is there any problem with the food quality? ",
    "How much satisfied are you about the delivery boy service? ",
    "How much satisfied are you about our overall service? ",
    "Please let me know, is there any issue with our service, application or any other problem? ",
    " END"
]

resList = [
  "ok sure",
  "It’s good",
  "No not at all, I like the food very much",
  "The delivery boy was well mannered, and well behaved, but took little time to understand my home location.",
  "I like your service",
  "Actually, the delivery took little bit of more time than expected. Otherwise, everything is fine",
  "  End list"
]

#list_of_score = [10]
list_of_score = ["" for x in range(len(list_of_questions))]
list_of_points =["" for x in range(len(list_of_questions))]
# Define conversation function
def feedback_bot():
    # Create conversation object
    conv = Conversation()

    # Start conversation
    #conv.add_user_input(input(
    print("Hello Good Morning! Hope you are doing good.\
    I am the feedback bot for the food that you ordered.\
    I would like to ask you three questions to get your feedback.\
     Hope you are fine with it.")


    ex = True
    Count =0
    RespCount=0
    # Loop until conversation is ended
    #while ex == True:
    for x in range(len(list_of_questions)):
        #ask question
        user_input_text = input(">> ")
        #user_input_text = resList[RespCount]
        RespCount +=1
        # Analyze sentiment of user input
        score = get_feedback_score(user_input_text)
        feedback_lvl = get_feedback_level(score)

        #print(f"user_input_text{user_input_text, }score : {score}, feedback lvl: {feedback_lvl}")
        #print(f"feedback lvl: {feedback_lvl}")
        # Determine response based on sentiment
        if feedback_lvl == 'Very Satisfied':
            res = np.random.choice(["Thanks for your response, "])
        elif feedback_lvl == 'Satisfied':
            res = np.random.choice(["Thank you for your positive feedback!, " ,"Thanks for your response, "])
        elif feedback_lvl == 'Neutral':
            res = np.random.choice(["Thank you for your positive feedback!, " ,"Thanks for your response, "])
        elif feedback_lvl == 'Dissatisfied':
            res = np.random.choice(["I'm sorry to hear that"])
        else: #"Very Dissatisfied"
            res = np.random.choice(["ohh i am extremelly sorry for that, "])

        # Print response
        print(f"{res}{list_of_questions[Count]}")
        list_of_score[Count] = feedback_lvl
        list_of_points[Count] = score
        Count +=1
        # End conversation if user input contains "bye"

        if "bye" in user_input_text.lower():
          ex =False
          print("Aborting...")

        if Count > 5:
            ex =False
            print("Thanks for your feedback")
            print(Count)
            print(list_of_points)

            AvgScore = sum(list_of_points)
            AvgScore = AvgScore / 6
            overall_Service_feedback = get_feedback_level(AvgScore)
            print(AvgScore)
            print("Summary: ")
            print(f"1.How much satisfied are you about the food quality? [[{list_of_score[2]}]]")
            print(f"2.How much satisfied are you about the delivery boy service? [[{list_of_score[3]}]]")
            print(f"3.How much satisfied are you about our overall service? [[{overall_Service_feedback}]]")

```
