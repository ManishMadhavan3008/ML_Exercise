```python
import numpy as np
import pandas as pd
from sklearn import linear_model as lm
import pickle as pick
```

```python
url = "https://raw.githubusercontent.com/ManishMadhavan3008/MachineLearning/main/canada_per_capita_income.csv"
df= pd.read_csv(url)
df
```

```python
reg = lm.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2018]])
```

```python
with open('model_pickel','wb') as f:
  pick.dump(reg,f)
```

```python
with open('model_pickel', 'rb') as f:
  BinModel = pick.load(f)
```

```python
BinModel.predict([[2022]])
```
