```python
import numpy as np
import pandas as pd
from  sklearn import linear_model as lm
import w2n2w
```

```python
# import data base url
url = "https://raw.githubusercontent.com/ManishMadhavan3008/ML_Exercise/main/2_LinearRegeresionMultiDimensional/hiring.csv"
db =  pd.read_csv(url)
db
```

```python
#clean data and
db.experience = db.experience.fillna('zero')
db.experience
```

lambda x: w2n.word_to_num(x): This is a lambda function that takes a single argument x (each element in the 'experience' column) and applies the word_to_num() function from the word2number library to convert the word to a number.

x represents each individual value in the 'experience' column.
w2n.word_to_num(x) converts the word to a number using the word2number library. For example, if x is 'two', it will be converted to 2.

```python

# conver NAn to negative one
db['experience']  = db['experience'].apply(lambda x: w2n2w.word_to_num(x))
db['experience']
#CnvrtedData = w2n.word_to_num(db.experience[])
```

```python
# correct 'test_score(out of 10)' column too with the maen value of math
import math

meanval= math.floor(db['test_score(out of 10)'].mean())
meanval
db['test_score(out of 10)'] =  db['test_score(out of 10)'].fillna(meanval)
db
```

```python
# after Cleaning of data need to call linear regression

reg = lm.LinearRegression()

reg.fit(db[['experience','test_score(out of 10)','interview_score(out of 10)']],db['salary($)'])

```

```python
exp = 7
score = 9
intScrore = 6

reg.predict([[exp,score,intScrore]])
```
